# The Neural Network Terminologies

### Neuron:
A neuron is an entity that takes values as inputs, performs operations on them and outputs the value.

### Input:
An input is an value that is passed on to the neuron for further processing.

### Weights:
Weights are the numeric values that helps neuron to understand how important an input is for the output,<br>
z=(x1*w1 + x2*w2 + ...)+b<br>
where, z=weighted sum,<br>
       x=input,<br>
       w=weights,<br>
       b=bias <br>
Weights are initially assigned randomly and are progressively improved.

### Bias
A bias is the value that helps the neuron to shift(move forward) the output even when the input does not provide any information, for example when output is zero,
you can see the bias as "b" in the formula above.

### Weighted Sum
Weighted sum (z) is the value calculated by the neuron before passing the "z" to the activation function,
don't worry we will learn the "activation function" further.

### Activation Function
Activation function is the function that activates the neuron to fire, here "to activate" mean to provide a output that is given as an final output or input to further neurons, the activation function decides wheather the neuron should act or not.
ex: sigmoid, <br>
Neuron fires strongly (output close to 1)



```
Inputs: x1, x2, x3
        │    │    │
        ▼    ▼    ▼
Weights: w1, w2, w3
        │    │    │
        └──► Multiply each input by its weight
                 │
                 ▼
              Add Bias (b)
                 │
                 ▼
      Weighted Sum (z = w1*x1 + w2*x2 + w3*x3 + b)
                 │
                 ▼
      Activation Function (e.g., sigmoid)
                 │
                 ▼
      Output (neuron firing strength)
```

### Loss
Loss is the mistakes done by the neural network<br>
loss=actual ouput / predicted output
Network would recover the loss by adjusting the weights as baises, we will see further how the network does that.

### Gradient
Gradient is an entity that decides the direction in which the weights and biases are to be adjusted to minimize the loss.<br>
Analogy

Think of a hill (loss landscape):<br>
Height = loss<br>
You are a hiker trying to reach the lowest point (minimum loss)<br>
Gradient = slope of hill at your position<br>
Positive slope → go down<br>
Negative slope → go down<br>
Step size → learning rate<br>

### Learning Rate
Learning Rate is the value that decides how fast a network should learn, fast learning--> less improvment<br>
slow learning--> better improvement<br>

### Back Propogation
Back propogation is the process which adjusts the weights and biases as decided by the gradient to minimize the loss of the network.

```
             ┌───────────┐
Inputs: x1,x2 ┤           │
             │ Neuron    │
Weights: w1,w2│           │
Bias: b       │           │
             └─────┬─────┘
                   │
                   ▼
        Weighted Sum (z = Σ(x*w) + b)
                   │
                   ▼
     Activation Function (sigmoid, ReLU, etc.)
                   │
                   ▼
       Output (Neuron firing strength)
                   │
                   ▼
       Compare with Actual → Compute Loss
                   │
                   ▼
              Backpropagation
                   │
                   ▼
          Compute Gradients (how to adjust)
                   │
                   ▼
       Update Weights & Bias (using learning rate)
                   │
                   ▼
        Next Forward Pass with updated weights
```
